```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,5])
y=np.array([3,8])
plt.plot(x,y,marker='s')
plt.plot(x,y,marker='*')
plt.plot(x,y,marker='^')
plt.plot(x,ymarker='o',ms=20,mec='r',mfc='g',ls=':',lw=5,c='y')#markersize # marker edge color(r,g,b,k,w,c,m,y) 
# marker face color # linestyle(:,--,-.,-) #line width 


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([2,5])
y1=np.array([3,8])
x2=[9,7,2]
y2=[5,9,12]
plt.grid(axis='x',ls='-.',lw=2,c='y')
f={'color':'brown','size':15, "family":'arial'}
plt.xlabel('x-values',fontdict=f)
plt.ylabel('y-values',fontdict=f)
plt.title('data analysis',loc='left',family='times new roman',color='cyan',size=20)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=[2,8]
y1=[3,12]
x2=[1,5,7]
y2=[1,5,10]
plt.plot(x1,y1,label='2023')
plt.plot(x2,y2,label='2024')
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=[2,8]
y1=[3,12]
x2=[1,5,7]
y2=[1,5,10]
plt.subplot(3,2,1)
plt.plot(x1,y1,'o-.b')
plt.title('d1')

plt.subplot(3,2,2)
plt.plot(x2,y2,'*:r')
plt.title('d2')

plt.subplot(3,2,3)
plt.plot(x1,y1,'o-.b')
plt.title('d3')

plt.subplot(3,2,4)
plt.plot(x1,x1,'*:r')
plt.title('d4')

plt.subplot(3,2,5)
plt.plot(x1,x1,'o-.b')
plt.title('d5')

plt.subplot(3,2,6)
plt.plot(y2,y2,'*:r')
plt.title('d6')

plt.show()
```
