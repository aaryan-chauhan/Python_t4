```python
import matplotlib.pyplot as plt
import numpy as np 
x1 = np.array([2,5])
y1 = np.array([3,8])
x2 = np.array([9,2])
y2 = np.array([5,9])
plt.subplot(3,2,1)
plt.plot(x1,y1,'*-.c')
plt.title('C')
plt.subplot(3,2,2)
plt.plot(x2,y2,'*-.r')
plt.title('R')
plt.subplot(3,2,3)
plt.plot(x1,y2,'*-.m')
plt.title('M')
plt.subplot(3,2,5)
plt.plot(x1,y2,'*-.g')
plt.title('G')
plt.subplot(3,2,4)
plt.plot(x2,y1,'*-.y')
plt.title('Y')
plt.subplot(3,2,6)
plt.plot(x2,y1,'*-.b')
plt.title('B')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
color = [0,10,20,30,35,40,45,50,60,65,70,90,100]
plt.scatter(x,y,c=color,cmap='viridis',s=[100],alpha=0.9)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
x = np.array(['A','B','C','D'])
y = np.array([100,200,50,150])
rgpo=['red','green','pink','orange']
lab = ['aaa','bbb','ccc','ddd']
plt.barh(x,y,color=['red','green','pink','orange'],height=0.2) # bar - vertical barh - horizontal
# plt.legend()
plt.xlabel("Company name")
plt.ylabel("Revenue")
plt.title("Comparition data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
x = [210,120,340,570,190]
plt.pie(x,labels=['AI','RAI','CSD','CE','IT'],colors=['r','g','b','c','m'],startangle=90,counterclock=False,radius=2,shadow=True,autopct='%1.f%%',explode=[0.2,0,0,0,0])
plt.show()
```

```python
# m (5,5) to (10,17) to (25,25) to (60,40) to (80,30) 
import matplotlib.pyplot as plt
import numpy as np 
x1 = np.array([5,5])
y1 = np.array([10,17])
x2 = np.array([25,25])
y2 = np.array([60,40])
x=[1,2,3,4,5]
Scores_men = (22, 30, 35, 35, 26)
Scores_women = (25, 32, 30, 35, 29)
plt.subplot(3,2,1)
plt.plot(x1,y1,'D:g')
plt.title('C')
plt.subplot(3,2,2)
plt.bar(x,Scores_men)
plt.bar(x,Scores_women)
plt.title('R')
plt.subplot(3,2,3)
plt.pie(x1,y2,'D:g')
plt.title('M')
plt.subplot(3,2,5)
plt.plot(x1,y2,'D:g')
plt.title('G')
plt.subplot(3,2,4)
plt.plot(x2,y1,'D:g')
plt.title('Y')
plt.subplot(3,2,6)
plt.plot(x2,y1,'D:g')
plt.title('B')
plt.show()
```
